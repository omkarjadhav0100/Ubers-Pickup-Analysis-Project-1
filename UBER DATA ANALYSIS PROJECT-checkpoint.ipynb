# 12. Price prediction function

```python
new_uber.head()
```

```python
def predict_price(name,source,surge_multiplier,icon):    
    loc_index = np.where(new_uber.columns==name)[0]

    x = np.zeros(len(new_uber.columns))
    x[0] = source
    x[1] = surge_multiplier
    x[2] = icon
    if loc_index >= 0:
        x[loc_index] = 1

    return random.predict([x])[0]
```

            Haymarket Square --> 5 , North End --> 6 , North Station --> 7 , Northeastern University --> 8 ,
